```python
import pandas as pd
```

```python
import os
```

```python
df = pd.read_csv('fighters.csv')
type(df)
pd.set_option('display.max_columns', None)
```

```python
#fill missing Record values
df.columns = df.columns.str.strip()
df['Place of Birth'] = df['Place of Birth'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Nickname'] = df['Nickname'].fillna('None')
df['Takedowns Landed Total'] = df['Takedowns Landed Total'].fillna('0')
df.head(15)
```

```python
record_split = df['Record'].str.extract(r'(\d+)-(\d+)-(\d+)')
record_index = df.columns.get_loc('Record')

df.insert(record_index + 1, 'Wins', record_split[0].fillna(0).astype(int))
df.insert(record_index + 2, 'Losses', record_split[1].fillna(0).astype(int))
df.insert(record_index + 3, 'Draws', record_split[2].fillna(0).astype(int))
```

```python
df = df.drop('Record', axis=1)
df.head(10)
```

```python
df['Nickname'] = df['Nickname'].str.replace('"', '').str.strip()
```

```python
def percent_to_float(val):
    try:
        return float(str(val).strip('%')) / 100
    except:
        return 0.0


df['Striking Accuracy'] = df['Striking Accuracy'].apply(percent_to_float)
df['Takedown Accuracy'] = df['Takedown Accuracy'].apply(percent_to_float)
df['Sig Str Def'] = df['Sig Str Def'].apply(percent_to_float)
df['Takedown Def'] = df['Takedown Def'].apply(percent_to_float)

df.head(15)
```

```python
import re

def extract_first_number(val):
    try:
        val = str(val).strip()
        # This finds the number BEFORE any space or parenthesis
        match = re.match(r'^(\d+)', val)
        if match:
            return int(match.group(1))
        else:
            return 0
    except:
        return 0

cols_to_clean = [
    'Sig Strikes While Standing',
    'Sig Strikes While Clinched',
    'Sig Strikes While Grounded',
    'Sig Strikes Head',
    'Sig Strikes Body',
    'Sig Strikes Leg',
    'Win by KO/TKO',
    'Win by Decision',
    'Win by Submission'
]
```

```python
for col in cols_to_clean:
    df[col] = df[col].apply(extract_first_number)
```

```python
def time_to_decimal_minutes(t):
    try:
        minutes, seconds = map(int, str(t).split(':'))
        return round(minutes + seconds / 60, 2)
    except:
        return 0.0
```

```python
col_index = df.columns.get_loc('Avg Fight Time')
df.insert(col_index + 1, 'Avg Fight Time (Minutes)', df['Avg Fight Time'].apply(time_to_decimal_minutes))
df = df.drop('Avg Fight Time', axis=1)
df.head()
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python
df['Takedowns Landed Total'] = df['Takedowns Landed Total'].astype(int)
```

```python
df.to_csv('cleaned_fighters.csv', index=False)
```
